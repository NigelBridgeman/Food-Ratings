## Build a lookup file using the postcode translation table as basis

## first, lets standadise the postocde field by removing spaces

```python
import pandas as pd

pcd = pd.read_csv("PCD.csv")
for index, row in pcd.iterrows():
    PostCode_S = row['pcd'].replace(" ","")
    pcd.loc[index, 'PostCode_S'] = PostCode_S
    pcd_sect_S = row['pcd_sect'].replace(" ","")
    pcd.loc[index, 'pcd_sect_S'] = pcd_sect_S
```

## get a list of all the postcode sectors

```python
## merge with our data file
```

```python
FR = pd.read_csv("FoodRating.csv")
# Only select records with rathing 0-5

FR = FR[(FR.RatingValue).isin(["0", "1", "2", "3", "4", "5"])]
FR = FR[(FR.PostCode) != "-"]
FR["RatingValue"]= FR["RatingValue"].astype('int')
```

```python
len(FR)
```

```python
pcd_FR = pd.merge(pcd, FR, how = "inner", on =['PostCode_S', 'PostCode_S'])
```

```python
len(pcd_FR)
```

```python
#Calculate the mean by PostCode Sector
```

```python
pcd_FR_mean_sect = pcd_FR.groupby(['pcd_sect'], sort=False)['RatingValue'].mean().reset_index()

pcd_FR_mean_sect2 = pcd_FR_mean_sect.rename(columns={'RatingValue': 'RVM_Sector'})

pcd_working = pd.merge(pcd, pcd_FR_mean_sect2, how = "left", on =['pcd_sect', 'pcd_sect'])
```

```python
len(pcd_working)
```

#Calculate the mean by PostCode Area

```python
pcd_FR_mean_area = pcd_FR.groupby(['pcd_area'], sort=False)['RatingValue'].mean().reset_index()

pcd_FR_mean_area = pcd_FR_mean_area.rename(columns={'RatingValue': 'RVM_Area'})

pcd_working = pd.merge(pcd_working, pcd_FR_mean_area, how = "left", on =['pcd_area', 'pcd_area'])
```

#Calculate the mean by PostCode District

```python
pcd_FR_mean_dist = pcd_FR.groupby(['pcd_dist'], sort=False)['RatingValue'].mean().reset_index()

pcd_FR_mean_dist = pcd_FR_mean_dist.rename(columns={'RatingValue': 'RVM_Dist'})

pcd_working = pd.merge(pcd_working, pcd_FR_mean_dist, how = "left", on =['pcd_dist', 'pcd_dist'])
```
